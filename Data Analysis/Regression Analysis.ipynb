```python
# Here we remane our columns to a list of shorter names we already difined above. WE do it by compy-pasting our dictionary.
        

dfr=DataFr.rename(columns={'1. What is your age?': '1_age', '2. Gender': '2_gender', '3. Relationship Status': '3_relationship_status', '4. Occupation Status': '4_occupation_status', '5. What type of organizations are you affiliated with?': '5_organization_affiliation', '6. Do you use social media?': '6_social_media_usage', '7. What social media platforms do you commonly use?': '7_social_media_platforms', '8. What is the average time you spend on social media every day?': '8_daily_social_media_time', '9. How often do you find yourself using Social media without a specific purpose?': '9_purposeless_social_media_usage_frequency', '10. How often do you get distracted by Social media when you are busy doing something?': '10_social_media_distraction_frequency', "11. Do you feel restless if you haven't used Social media in a while?": '11_social_media_restlessness', '12. On a scale of 1 to 5, how easily distracted are you?': '12_distraction_level', '13. On a scale of 1 to 5, how much are you bothered by worries?': '13_worry_level', '14. Do you find it difficult to concentrate on things?': '14_concentration_difficulty', '15. On a scale of 1-5, how often do you compare yourself to other successful people through the use of social media?': '15_social_media_comparison_frequency', '16. Following the previous question, how do you feel about these comparisons, generally speaking?': '16_comparison_feelings', '17. How often do you look to seek validation from features of social media?': '17_social_media_validation_seeking', '18. How often do you feel depressed or down?': '18_depression_frequency', '19. On a scale of 1 to 5, how frequently does your interest in daily activities fluctuate?': '19_activity_interest_fluctuation', '20. On a scale of 1 to 5, how often do you face issues regarding sleep?': '20_sleep_issue_frequency'})
```

```python
# we devide our list of columns to two kind one our dependent variable and the other indpendent variable
# in order to facilitate a function that can run regression model for we define a dictionary (DctA)

ListIn=['8_daily_social_media_time',
       '9_purposeless_social_media_usage_frequency',
       '10_social_media_distraction_frequency', '11_social_media_restlessness']
List=['a','b','c','d','e','f','g','h','j']
LsCo=['A','B','C','D','E','F','G','H','J']
pr=  ['k','l','m','n','o','p','q','r']
rsq= ['s','t','u','v','w','x','y','z']
ListV= ['12_distraction_level', '13_worry_level', '14_concentration_difficulty',
       '15_social_media_comparison_frequency', '16_comparison_feelings',
       '17_social_media_validation_seeking', '18_depression_frequency',
       '19_activity_interest_fluctuation', '20_sleep_issue_frequency']
DctA={}
for s in range(len(List)):
    for d in range(len(LsCo)):
        for t in range(len(ListV)):
            for n in range(len(pr)):
                for m in range(len(rsq)):
                    if s==d==t:
                        DctA[ListV[t]]=LsCo[d]
                        DctA[ListV[t]]= [List[s], LsCo[d], pr[n], rsq[m]]
            
print(DctA)
```

```python
# Here we are transfering hour spent on social media to average hour ( we select 6 for more than 5 hours)

dfr.replace({'8_daily_social_media_time':{'More than 5 hours': 6,
                                              'Between 2 and 3 hours': 2.5,
                                              'Between 3 and 4 hours': 3.5,
                                              'Between 1 and 2 hours': 1.5,
                                              'Between 4 and 5 hours': 4.5,
                                              'Less than an Hour': 0.5}}, inplace=True)
```

```python
# We run regression to and finding its coeffiecient and intercept for a range of dependent and independent variabel.
# We define this function to run regression to multip


def LinReg(L, Dict):
    
    '''The function LinReg is will run regression of a Ditionary (Dict) keys on
    a indepent variable of elements of a List (L)
    L: a list of independent variable
    Dict: a dictionary of depenent variable each key with four values
    Output: printing intercept, coefficient and R-square'''
    
    model=LinearRegression()
    for e in L:
        print('Independent:',e)
        print('-----------------------------')
        for i,j in Dict.items():
            
            model.fit(dfr[[e]], dfr[[i]])
            j[0]=model.intercept_
            j[1]=model.coef_
            print('Intercentp',i, ':', j[0], 'Coefficeint',i,':', j[1])
            j[2]=model.predict(dfr[[e]])
            j[3]=r2_score( dfr[[i]]
                              , j[2])

            print('R-Squre',i, ':',j[3], end='\n\n')
        

print(LinReg(ListIn, DctA))
```

```python
# We would like to have ANOVA table and check if they are statistically significant
# We creat a matrix of indpendent variable as X
# We also add constant to the matrix to count for intercept and assigning it Xa
# We defind our dependent variable as y

X=dfr.loc[:,['8_daily_social_media_time',
       '9_purposeless_social_media_usage_frequency',
       '10_social_media_distraction_frequency', '11_social_media_restlessness']]
Xa=sm.add_constant(X)

y=dfr.loc[:,['12_distraction_level']]
```

```python
# WE run our linear regression model and print out the summary of results

lm=sm.OLS(y, Xa).fit()
lm
print(lm.summary())
```

```python
# We modify our matrix of indpendent variable as X1
# We also add constant to the matrix to count for intercept and assigning it Xa1

X1=dfr.loc[:,[
       
       '10_social_media_distraction_frequency', '11_social_media_restlessness']]
Xa1=sm.add_constant(X1)
```

```python
# WE run our linear regression model and print out the summary of results

linM=sm.OLS(y, Xa1).fit()
linM
print(linM.summary())
```

```python
# WE run our linear regression model using a function and iterating over the dependent variables.


def Reg(y,x):
    '''The function Reg is running linear regression
    y: is dictionary of dependent variable
    x: is matrix of independent variable with constant
    output: prints summary of each key in the dictionary'''
    for i in y:
        lm=sm.OLS(dfr[i],x).fit()

        print(lm.summary())
Reg(DctA, Xa)
```

```python
# Calling our Reg funtion on only two indpendent variables

Reg(DctA, Xa1)
```

```python
# Redefining our matrix of indpendent variables including three of them
# Calling our Reg funtion on three indpendent variables

X2=dfr.loc[:,[ '9_purposeless_social_media_usage_frequency',
       '10_social_media_distraction_frequency', '11_social_media_restlessness']]
Xa2=sm.add_constant(X2)
Reg(DctA, Xa2)
```
